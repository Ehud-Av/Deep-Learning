$$

#### Question 2: Properties of Convolution

$$
f \star g = g \star f.
$$
Proof
$$
f \star g (t) = \int_{-\infty}^{\infty} f(s) \ g(t-s) ds = \{  s = t - \hat s, ds = -d \hat s \} =
\int_{\infty}^{-\infty} f(t- \hat s) \ g(\hat s) - d \hat s =
$$
$$
=\int_{-\infty}^{\infty} f(t- \hat s) \ g(\hat s) d \hat s 

=g \star f.

$$
End of proof.

**b. Distributivity**: 
$$
f \star (\alpha g + \beta h) 
= \int_{-\infty}^{\infty} f(s) \ (\alpha g + \beta h)(t-s) ds 
=\int_{-\infty}^{\infty} f(s) \ (\alpha g(t-s) + \beta h (t-s)) ds=
$$
$$
= \alpha \int_{-\infty}^{\infty} f(s) \ g(t-s) ds +\beta \int_{-\infty}^{\infty} f(s) \ h (t-s)ds
=\alpha (f \star g) + \beta (f \star h),
$$  
End of proof. 

Assumpotions: 
1) We work on a linear space in which: <br>
$(\alpha g)(t-s) = \alpha g(t-s)$, \
$(g + h)(t-s) = g(t-s) +  h (t-s)$ 
2) $ \alpha, \beta \in \mathbb{R} $.

**a. Derivative of Convolution**: Derive the expression for the derivative of the convolution $ f(x) \star g(x) $:  
$$
\frac{d}{dx} \big(f(x) \star g(x)\big).
$$  
Proof:
$$
\frac{d}{dx} \big(f(x) \star g(x)\big) =  \frac{d}{dx} \int_{-\infty}^{\infty} f(s) \ g(x-s) ds =
$$
$$
=\int_{-\infty}^{\infty} \frac{d}{dx} f(s) \ g(x-s) ds = \int_{-\infty}^{\infty} f(s) \ g'(x-s) ds 
= \big(f(x) \star g'(x)  \big) = \big(f \star \frac{dg}{dx}  \big).
$$
$$
\text{ In the same way:}
$$
$$
\frac{d}{dx} \big(f(x) \star g(x)\big) = \frac{d}{dx} \big(g(x) \star f(x)\big) 
= \big(g \star \frac{df}{dx}  \big) = \big(\frac{df}{dx} \star g \big).
$$
$$
\text{ So in conclusion, we got: }
$$
$$
\frac{d}{dx} \big(f(x) \star g(x)\big) =  \big(f(x) \star \frac{d g}{dx}\big) =  \big(\frac{df}{dx} \star g(x)\big)
$$
Let's verify this property for specific functions:
$$
g(x) = x^2
$$
$$
g'(x)=2x

$$
f(x)=

$$

**b. Inverse of Convolution**: Discuss the conditions under which the convolution $ f(x) \star g(x) $ can be inverted. Derive the theoretical framework for deconvolution or solving for $ f(x) $ or $ g(x) $ given their convolution.

To solve that, we will need to use fourier transform, therefore we assume that there it is a fourier transform for f(x), g(x) and its convolution.
Let's solve for $f(t)$. 
Let's notate: 
$$
h(t) = f(t) \star g(t)
$$
Let's apply fourier transform.
$$
\mathcal{F}(f(t) \star g(t)) = \mathcal{F} (h(t))
$$
Let us use the convolution property of Fourier transform.
$$
\mathcal{F}(f(t)) \mathcal{F}(g(t)) = \mathcal{F}(h(t))
$$
Therefore, assuming $\mathcal{F}(g(t)) \neq 0$
$$
\mathcal{F}(f(t)) = \frac{\mathcal{F}(h(t))}{\mathcal{F}(g(t))}
$$
We will apply inverse fourier transform.
$$
\mathcal{F^{-1}}(\mathcal{F}(f(t))) = \mathcal{F^{-1}}(\frac{\mathcal{F}(h(t))}{\mathcal{F}(g(t))})
$$
And we get,
$$
f(t) = \mathcal{F^{-1}}(\frac{\mathcal{F}(h(t))}{\mathcal{F}(g(t))})
$$
In the same way
$$
g(t) = \mathcal{F^{-1}}(\frac{\mathcal{F}(h(t))}{\mathcal{F}(f(t))})
$$

So we got:
$$
f(t) = \mathcal{F^{-1}}(\frac{\mathcal{F}(h(t))}{\mathcal{F}(g(t))})
$$
$$
g(t) = \mathcal{F^{-1}}(\frac{\mathcal{F}(h(t))}{\mathcal{F}(f(t))})
$$
